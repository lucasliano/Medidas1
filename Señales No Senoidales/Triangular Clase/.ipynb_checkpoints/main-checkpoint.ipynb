```python
import numpy as np
from scipy import signal
from matplotlib import pyplot as plt 
```

```python
CANT_CICLOS = 8
duty = 0.5

Vp = 3.08
Vn = -1.08
Vpp = Vp - Vn
Vdc = (Vp + Vn)/2

Amplitud = (Vpp/2)

```

```python
x = np.linspace(0, CANT_CICLOS*2*np.pi, 4000, endpoint = False) 
y = Amplitud * signal.sawtooth(x, duty) + Vdc

plt.plot(x, y)
```

## Cálculo de valor $V_{med}$

```python
# Nota: Podriamos calcular el valor de Vdc como np.mean(y), 
#       comopero esto solo funcionará mientras que CANT_CICLOS sea un entero
print("Valor DC: ",np.mean(y))
print("Valor DC: ",Vdc)
```

## Cálculo de valor $V_{|me|}$
Ahora tenemos que trabajar pensando que la señal se va a acoplar con el capacitor en el modo AC.

Por lo que al calcular el $V_{|me|}$ lo único que debemos hacer es calcular un promedio temporal de esa señal. 

\begin{equation}
    V_{|me|} = \frac{1}{T} \int_{0}^{T} V(t) dt
\end{equation}

Que para este caso particular, podemos pensar la integral como que es dos veces el área de un triángulo.

\begin{equation}
    V_{|me|} = \frac{2}{T} (\frac{\frac{T}{2} \frac{Vpp}{2}}{2}) = \frac{Vpp}{4}
\end{equation}


```python
# Simulamos el acople
y_ac = np.abs(y-Vdc)
plt.plot(x, y_ac)
```

```python
# Calculamos analiticamente el Vme y anotamos el Kme
Kme = np.pi / (2*np.sqrt(2))
Vme = Amplitud / 2
Vi_me = Vme * Kme

print("Valor medio de modulo: ",np.mean(y_ac))
print("Valor medio de modulo: ",Vme)
print("Valor observado: ", Vi_me)
```

## Cálculo de valor $V_{e}$ (AC + DC)


Por lo que al calcular el valor RMS (AC+DC) debemos calcularlo como

\begin{equation}
    V_{e} (AC + DC) = \sqrt{V_{e_{AC}}^{2} + V_{dc}^{2} }
\end{equation}

Luego, 

\begin{equation}
    V_{e_{AC}} = \dfrac{\frac{Vpp}{2}}{\sqrt{3}}
\end{equation}

```python
valor_rms = np.sqrt((Vpp/2)/(np.sqrt(3))**2 + Vdc**2) 
print("Valor RMS (AC+DC): ", valor_rms)
```
