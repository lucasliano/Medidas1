
# SENOIDAL RECTIFICADA MEDIO CICLO #####

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
amplitud=8.56
t=np.linspace(0,2*np.pi,10000,endpoint=True)
v=amplitud*np.sin(10*t)
plt.plot(t,v)
print("el valor RMS de la senoidal es = {}".format(np.sqrt(np.mean(np.power(v,2)))))
```

```python
### Aprovecho para mostrar una funcionalidad muy copada ###

indices = v > 0 # Es un vector booleano que voy a usar para direccionar 

rectificada = v * indices
plt.plot(t,rectificada)279785994649733
```

### Hasta aca hemos generado la señal que debe medir el tester

```python
print("el valor RMS de la senoidal recortada es {}".format(np.sqrt(np.mean(np.power(rectificada,2)))))
```

Lo importante es que con el fluke189 podemos medir el valor AC+DC como con el osciloscopio acoplado en DC

### El tester en AC

```python
print ("el tester tiene acople capacitivo, por lo tanto genera una senal de valor medio 0")
print("el valor medio de la señal de entrada es {}".format(np.mean(rectificada)))
print("la teoria dice que seria Vp/pi = {}".format(amplitud/np.pi))
rectificada_despues_C = rectificada-np.mean(rectificada)
print("despeus del capacitor queda en {}".format(round(np.mean(rectificada_despues_C)),4))
```

#### El valor de DC o valor medio algebraico de la señal se puede obtener poniendo el tester en DC

<img src=".\DSC_0211.jpg">

```python
plt.plot(t,rectificada_despues_C,t,np.zeros(len(t)))
```

### Luego del capacitor el tester tiene un rectificador de onda completa

```python
rectificada_despues_C_despues_rect=np.abs(rectificada_despues_C)
plt.plot(t,rectificada_despues_C_despues_rect,t,np.zeros(len(t)))
```

```python
print("el valor medio de modulo que detecta el tester es {}".format(np.mean(rectificada_despues_C_despues_rect)))
print("que despues multiplica por su constante de calibracion pi/2raiz(2)  \n El tester indica {}".format(np.mean(rectificada_despues_C_despues_rect)*np.pi/(2*np.sqrt(2))))
```
